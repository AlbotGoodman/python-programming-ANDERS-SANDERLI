## Importing and running the script

```python
# import lab3 as l3
from lab3 import read_file, classifier, write_file, create_figure, plot_line, plot_scatter
import numpy as np
import matplotlib.pyplot as plt

# default line values
K = -1
X = np.linspace(-10, 10, 600)
M = 0

# read the data
dataframe = read_file("unlabelled_data.csv")
labels = classifier(dataframe, K, M)
dataframe_labelled = write_file(dataframe, labels)
ax = create_figure(dataframe)

# plot the default graph
plot_line(ax, X, K, M)
plot_scatter(ax, dataframe_labelled)
plt.show()
```

As you can tell from the above, a very simple decision boundary clearly splits the differenct data points in two clusters.  

To get this line I took the easy route of just going with simply values for *K* and *M*.
